```python
#Delete useless columns 
df = df.drop(columns=['fiscal_month_number', 'fiscal_year','fiscal_year', 'same_day_previous_year', 'same_day_previous_year_key'])
```

```python
# Replace NaN in holiday_name by NoHoliday 
df['holiday_name'].fillna('NoHoliday', inplace=True)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['week_begin_date'] = pd.to_datetime(df['week_begin_date'])
```

```python
df.dtypes
```

```python
df.rename(columns={'date_key':'date_id', 'week_begin_date_key':'week_begin_date_id'}, inplace=True)
```

```python
df.insert(0, "date_key",  np.arange(len(df)))
```
